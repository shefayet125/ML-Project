# <b><b>Feature Engineering and Model Evaluation

```python
def outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1 - 1.5 *IQR
    upper_bound = Q3 + 1.5 * IQR
    return (df < lower_bound) | (df > upper_bound)
```

```python
detect_outliers = outliers(df)
detect_outliers.sum().sort_values(ascending=False)
```

```python
for col in df.columns:
    fig,(ax1,ax2) = plt.subplots(1,2,figsize =(12,4))
    sns.boxplot(df[col],ax=ax1)
    ax1.set_title(f'Box Plot of {col}')
    ax2.scatter(range(len(df)),df[col],c=['blue' if not x else 'red' for x in detect_outliers[col]])
    ax2.set_title(f'Scatter Plot of {col} with Outliers Highlighted')
    plt.tight_layout()
    plt.show()
```

```python
#Standardized data for ruduce impact of outlier.
df = (df - df.mean() / df.std())
df.head()
```

```python
fig,axes = plt.subplots(5,4,figsize=(15,20))
axes = axes.flatten()
for i,s in enumerate(df.columns):
    axes[i].hist(df[s],bins=20)
    axes[i].set_title(s)

plt.tight_layout()
fig.suptitle('Check All the Features Distribution',fontsize=25,y=1.03)
plt.show()
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12,6))
sns.heatmap(corr_matrix,annot=True,cmap='viridis',fmt='.2f')
plt.title('Correlation Plot',fontsize=20)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
import xgboost as xgb
from xgboost import XGBRegressor
from xgboost import plot_importance
```

```python
X = df.drop(columns='price',axis=1)
y= df[['price']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr_predict = lr.predict(X_test)
```

```python
R_Square = r2_score(y_test,lr_predict)
AME = mean_absolute_error(y_test,lr_predict)
RMSE = np.sqrt(mean_squared_error(y_test,lr_predict))
MAPE = '{:.2f}%'.format(mean_absolute_percentage_error(y_test,lr_predict))

series= pd.Series({'Lr R-Square':R_Square,'Lr AME':AME,'Lr RMSE':RMSE,'Lr MAPE':MAPE})
metric = pd.DataFrame(series,columns=['Value'])
metric.index.name = 'Accuracy Metrics'
metric.T
```

## Featre selection by Embedded Method 

```python
feature_rfe = RFE(
    RandomForestRegressor(n_estimators=15, random_state=0),
    n_features_to_select=10,
    step=2)

feature_rfe.fit(X_train, y_train)
```

```python
selected_feature = X_train.columns[(feature_rfe.get_support())]
print(f'Number of selected feature: {len(selected_feature)}')
print(selected_feature)
```

```python
RFE_predict = feature_rfe.predict(X_test)
```

```python
R_Square = r2_score(y_test,RFE_predict)
AME = mean_absolute_error(y_test,RFE_predict)
RMSE = np.sqrt(mean_squared_error(y_test,RFE_predict))
MAPE = '{:.2f}%'.format(mean_absolute_percentage_error(y_test,RFE_predict))

series= pd.Series({'RFE R-Square':R_Square,'RFE AME':AME,'RFE RMSE':RMSE,'RFE MAPE':MAPE})
metric = pd.DataFrame(series,columns=['Value'])
metric.index.name = 'Accuracy Metrics'
metric.T
```

## Feature selection by XgBoost

```python
xgb_model = xgb.XGBRegressor()
xgb_model.fit(X_train, y_train)
plot_importance(xgb_model,max_num_features=10,importance_type='gain')
plt.show()
```

<b><b>Comment:
* Feature selection in xgboost there are a problem in the parameter `importance_type`, there 3 types of input that are `cover, gain,weight.` So, each types selected features are different and it's too much hard to chose the appropiate types for a model.

```python
from sklearn.model_selection import RandomizedSearchCV

parameter_tune = {
    'n_estimators': [100, 300, 500],           # Number of trees
    'learning_rate': [0.01, 0.05, 0.1],        # Step size shrinkage
    'max_depth': [3, 5, 7],                    # Tree depth
    'min_child_weight': [1, 3, 5],             # Minimum child weight
    'subsample': [0.6, 0.8, 1.0],              # Fraction of training instances
    'colsample_bytree': [0.6, 0.8, 1.0],       # Fraction of features
    'gamma': [0, 1, 5],                        # Minimum loss reduction
    'alpha': [0, 0.1, 1],                      # L1 regularization
    'lambda': [0, 1]                           # L2 regularization
}

# Perform random search
random_search = RandomizedSearchCV(estimator=xgb_model, param_distributions=parameter_tune, 
                                   n_iter=50, scoring='neg_mean_squared_error', 
                                   cv=3, verbose=2, n_jobs=-1, random_state=42)

# Fit the random search
random_search.fit(X_train, y_train)

# Best parameters and score
print(f"Best Parameters: {random_search.best_params_}")
print(f"Best Score: {-random_search.best_score_}")
```

```python
best_xgb_model = random_search.best_estimator_
```

```python
y_pred_xgb = best_xgb_model.predict(X_test)
R_Square = r2_score(y_test,y_pred_xgb)
AME = mean_absolute_error(y_test,y_pred_xgb)
RMSE = np.sqrt(mean_squared_error(y_test,y_pred_xgb))
MAPE = '{:.2f}%'.format(mean_absolute_percentage_error(y_test,y_pred_xgb))

series= pd.Series({'Xgb R-Square':R_Square,'Xgb AME':AME,'Xgb RMSE':RMSE,'Xgb MAPE':MAPE})
metric = pd.DataFrame(series,columns=['Value'])
metric.index.name = 'Accuracy Metrics'
metric.T
```

## Feature selection by SHAP

```python
import shap
shap.initjs()
best_xgb_model.fit(X_test,y_test)
```

```python
# Initialize the SHAP explainer with your XgBoost model
explainer = shap.TreeExplainer(best_xgb_model)

# Calculate SHAP values for the entire test set
shap_values = explainer(X_test)

# Plot the summary plot for the SHAP values
shap.summary_plot(shap_values, X_test,plot_size=(10,5))
```

```python
#the JavaScript visualizations are initialized
shap.initjs()

# Plot the SHAP values for the 10th observation
shap.force_plot(
    explainer.expected_value,     # The model's expected value (baseline prediction)
    shap_values.values[10, :],    # SHAP values for the 10th observation
    X_test.iloc[10, :]            # The feature values for the 10th observation
)
```

```python
# Plot the SHAP values as a bar plot with the top 15 features.
shap.summary_plot(
    shap_values.values,
    X_test,                      
    max_display=15,            
    show=False,
    plot_type='bar',             
    feature_names=X_test.columns,
    plot_size=(10,5)            
)
```
